# **BERT Model: Tutorial**

```python
import tensorflow_hub as hub
import tensorflow_text as text

preprocess_url = "https://tfhub.dev/tensorflow/bert_en_uncased_preprocess/3"
encoder_url = "https://tfhub.dev/tensorflow/bert_en_uncased_L-12_H-768_A-12/4"
```

```python
bert_preprocess_model = hub.KerasLayer(preprocess_url)
```

```python
text_test = ['nice movie indeed','I love python programming']
text_preprocessed = bert_preprocess_model(text_test)
text_preprocessed.keys()
```

```python
text_preprocessed['input_mask']
```

```python
text_preprocessed['input_type_ids'], text_preprocessed['input_word_ids']
```

```python
bert_model = hub.KerasLayer(encoder_url)
bert_result = bert_model(text_preprocessed)
bert_result.keys()
```

```python
bert_result['sequence_output'], bert_result['pooled_output']
```

```python
len(bert_result['encoder_outputs'])
```

```python
bert_result['encoder_outputs'][-1] == bert_result['sequence_output']
```

```python
from transformers import pipeline

emotional_model = pipeline('text-classification', model='SamLowe/roberta-base-go_emotions')
label = emotional_model(['I am not having a great day.',"i'm sorry that the order got delayed"])
label
```

```python
import pandas as pd

df = pd.read_csv('/content/eng_dataset.csv')
df.columns = ['ID', 'Emotion', 'Text']
df.head()
```

```python
simple_text = df[:200]
simple_text.shape, df.shape
```

```python
def get_emotion(text):
  return emotional_model(text)[0]['label']

simple_text['Text'][:10].apply(get_emotion)
```

```python
simple_text['Emotion'] = simple_text['Text'].apply(get_emotion)
simple_text.head()
```
