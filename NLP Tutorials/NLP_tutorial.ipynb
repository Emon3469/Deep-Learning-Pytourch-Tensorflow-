```python
import re

chat1='codebasics: Hello, I am having an issue with my order # 412889912'
chat2='codebasics: I have a problem with my order number 412889912'
chat3='codebasics: My order 412889912 is having an issue, I was charged 300$ when online it says 280$'

pattern = 'order[^\d]*(\d*)'
matches1 = re.findall(pattern, chat1)
matches2 = re.findall(pattern, chat2)
matches3 = re.findall(pattern, chat3)

matches = matches1 + matches2 + matches3
matches, matches1, matches2, matches3
```

```python
def get_pattern_match(pattern, text):
  matches = re.findall(pattern, text)
  if matches:
    return matches[0]
```

```python
chat1 = 'codebasics: you ask lot of questions 😠  1235678912, abc@xyz.com'
chat2 = 'codebasics: here it is: (123)-567-8912, abc@xyz.com'
chat3 = 'codebasics: yes, phone: 1235678912 email: abc@xyz.com'

pattern = '[a-z0-9A-Z_]*@[a-z0-9A-Z]*\.[a-zA-Z]*'
matches1 = get_pattern_match(pattern, chat1)
matches2 = get_pattern_match(pattern, chat2)
matches3 = get_pattern_match(pattern, chat3)

matches1, matches2, matches3
```

```python
def extract_pettern_info(text):
  age = get_pattern_match('age (\d+)', text)
  full_name = get_pattern_match('Born(.*)\n', text)
  birth_date = get_pattern_match('Born.*\n(.*)\(age', text)
  birth_place = get_pattern_match('\(age.*\n(.*)', text)
  return {
    'age': int(age),
    'name': full_name.strip(),
    'birth_date': birth_date.strip(),
    'birth_place': birth_place.strip()
  }
```

```python
from nltk.tokenize import word_tokenize

transactions = "Tony gave two $ to Peter, Bruce gave 500 € to Steve"
words = word_tokenize(transactions)

num_words = {'one', 'two', 'three', 'four', 'five'}
currencies = {'$', '€', '£', '¥'}

for i in range(len(words) - 1):
  word = words[i]
  next_word = words[i + 1]
  if (word.isdigit() or word.lower() in num_words) and next_word in currencies:
    print(f"{word} {next_word}")
```

```python
text = '''Ravi and Raju are the best friends from school days.They wanted to go for a world tour and
visit famous cities like Paris, London, Dubai, Rome etc and also they called their another friend Mohan to take part of this world tour.
They started their journey from Hyderabad and spent next 3 months travelling all the wonderful cities in the world and cherish a happy moments!
'''
```

```python
import spacy

nlp = spacy.load('en_core_web_sm')
doc = nlp(text)

proper_noun = []
count = 0

for word in doc:
  if word.pos_ == 'PROPN':
    proper_noun.append(word.text)
    count += 1
proper_noun, count
```

```python
text = '''The Top 5 companies in USA are Tesla, Walmart, Amazon, Microsoft, Google and the top 5 companies in
India are Infosys, Reliance, HDFC Bank, Hindustan Unilever and Bharti Airtel'''

doc = nlp(text)
company_names = []

for ent in doc.ents:
  if ent.label_ == 'ORG':
    company_names.append(ent)
company_names, len(company_names)
```
