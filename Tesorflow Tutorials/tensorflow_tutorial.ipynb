```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import numpy as np

(X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()
len(X_train), len(X_test), X_train[0].shape, X_train[0]
```

```python
plt.matshow(X_train[0])
```

```python
X_train = X_train / 255
X_test = X_test / 255
X_train[0], y_train
```

```python
X_train_flattenend = X_train.reshape(len(X_train), 28*28)
X_test_flattenend = X_test.reshape(len(X_test), 28*28)

X_train_flattenend.shape, X_train_flattenend[0]
```

```python
model = keras.Sequential([
    keras.layers.Dense(10, input_shape=(784,), activation='sigmoid')
])
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(X_train_flattenend, y_train, epochs=5)
```

```python
model.evaluate(X_test_flattenend, y_test)
```

```python
y_predicted = model.predict(X_test_flattenend)
y_predicted
```

```python
plt.matshow(X_test[0])
```

```python
np.argmax(y_predicted[0])
```

```python
y_predicted_label = [np.argmax(i) for i in y_predicted]
y_predicted_label[:5]
```

```python
cm = tf.math.confusion_matrix(labels=y_test, predictions=y_predicted_label)
cm
```

```python
import seaborn as sns
plt.figure(figsize = (10, 7))
sns.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
(X_train, y_train), (X_test, y_test) = keras.datasets.cifar10.load_data()
X_train.shape, y_train.shape
```

```python
def plot_sample(idx):
  plt.figure(figsize=(10, 1))
  plt.imshow(X_train[idx])
plot_sample(0)
```

```python
classes = ["airplane","automobile","bird","cat","deer","dog","frog","horse","ship","truck"]
classes[y_train[3][0]], y_train[:3], y_test.shape, X_train.shape
```

```python
X_train_scaled = X_train / 255
X_test_scaled = X_test / 255

y_train_categorical = keras.utils.to_categorical(y_train, num_classes=10)
y_test_categorical = keras.utils.to_categorical(y_test, num_classes=10)
y_train[0:5], y_train_categorical[0:5]
```

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(32, 32, 3)),
    keras.layers.Dense(3000, activation='relu'),
    keras.layers.Dense(1000, activation='relu'),
    keras.layers.Dense(10, activation='sigmoid')
])

model.compile(optimizer='SGD', loss='categorical_crossentropy', metrics=['accuracy'])
model.fit(X_train_scaled, y_train_categorical, epochs=1)
```

```python
np.argmax(model.predict(X_test_scaled)[0])
y_test[0]
```

```python
def get_model():
  model = keras.Sequential([
    keras.layers.Flatten(input_shape=(32, 32, 3)),
    keras.layers.Dense(3000, activation='relu'),
    keras.layers.Dense(1000, activation='relu'),
    keras.layers.Dense(10, activation='sigmoid')
  ])

  model.compile(optimizer='SGD', loss='categorical_crossentropy', metrics=['accuracy'])
  return model
```

```python
%%timeit -n1 -r1
with tf.device('/CPU:0'):
  cpu_model = get_model()
  cpu_model.fit(X_train_scaled, y_train_categorical, epochs=1)
```

```python
%%timeit -n1 -r1
with tf.device('/GPU:0'):
  cpu_model = get_model()
  cpu_model.fit(X_train_scaled, y_train_categorical, epochs=1)
```

```python
(train_images, train_labels), (test_images, test_labels) = keras.datasets.fashion_mnist.load_data()
```

```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```python
plt.imshow(train_images[0])
```

```python
train_images, train_labels[0], train_labels[0], class_names[train_labels[0]]
```

```python
plt.figure(figsize=(3, 3))
for i in range(5):
  plt.imshow(train_images[i])
  plt.xlabel(class_names[train_labels[i]])
  plt.show()
```

```python
train_images_scaled = train_images / 255.0
test_images_scaled = test_images / 255.0

def get_model(hidden_layers=1):
  layers = [keras.layers.Flatten(input_shape=(28, 28))]
  for i in range(hidden_layers):
    layers.append(keras.layers.Dense(500, activation='relu'))
  layers.append(keras.layers.Dense(10, activation='sigmoid'))
  model = keras.Sequential(layers)
  model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
  return model

model = get_model(1)
model.fit(train_images_scaled, train_labels, epochs=5)
```

```python
model.predict(test_images_scaled)[2], test_labels[2]
```

```python
%%timeit -n1 -r1
with tf.device('/GPU:0'):
  cpu_model = get_model(1)
  cpu_model.fit(train_images_scaled, train_labels, epochs=5)
```

```python
%%timeit -n1 -r1
with tf.device('/GPU:0'):
  cpu_model = get_model(5)
  cpu_model.fit(train_images_scaled, train_labels, epochs=5)
```
