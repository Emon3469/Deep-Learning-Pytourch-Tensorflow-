Arithmetic Operations on Images using OpenCV

```python
import matplotlib.pyplot as plt
import numpy as np

img1 = cv.imread('/content/lambo.png')
img2 = cv.imread('/content/robot.png')

if img1.shape != img2.shape:
  img2 = cv.resize(img2, (img1.shape[1], img1.shape[0]))

line_thick = 5
height = img1.shape[0]
line = np.full((height, line_thick, 3), (0, 0, 255), dtype=np.uint8)

side_by_side = np.hstack((img1, line, img2))
side_by_side_rgb = cv.cvtColor(side_by_side, cv.COLOR_BGR2RGB)

plt.figure(figsize=(12, 6))
plt.imshow(side_by_side_rgb)
plt.title('input1 input2')
plt.axis('off')
plt.show()
```

```python
def plot_fig(operation, title_text):
  plt.figure(figsize=(12, 6))
  plt.imshow(operation)
  plt.title(title_text)
  plt.show()
```

```python
added = cv.add(img1, img2)
added_rgb = cv.cvtColor(added, cv.COLOR_BGR2RGB)

plot_fig(added_rgb, 'Added Images')
```

```python
weighted = cv.addWeighted(img1, 0.7, img2, 0.3, 0)
weighted_rgb = cv.cvtColor(weighted, cv.COLOR_BGR2RGB)

plot_fig(weighted_rgb, 'Weighted Images')
```

```python
sub = cv.subtract(img1, img2)
sub_rgb = cv.cvtColor(sub, cv.COLOR_BGR2RGB)

plot_fig(sub_rgb, 'Subtracted Images')
```

```python
And = cv.bitwise_not(img1, img2)
And_rgb = cv.cvtColor(And, cv.COLOR_BGR2RGB)

plot_fig(And_rgb, 'Bitwise And')
```

```python
width, height = 250, 350
img = cv.imread('/content/cards.jpg')

pts1 = np.float32([[752,118],[1120,265],[540,668],[871,830]])
pts2 = np.float32([[0,0], [width, 0], [0,height], [width, height]])

metrix = cv.getPerspectiveTransform(pts1, pts2)
imgOutput = cv.warpPerspective(img, metrix, (width, height))

cv2_imshow(img)
cv2_imshow(imgOutput)
```

```python
weight, height = 250 , 350
img = cv.imread('/content/docs.jpg')

pst1 = np.float32([[307, 250], [751, 272], [213, 859], [711, 866]])
pst2 = np.float32([[0, 0], [weight, 0], [0, height], [weight, height]])

metrix = cv.getPerspectiveTransform(pst1, pst2)
img_out = cv.warpPerspective(img, metrix, (weight, height))

cv2_imshow(img)
cv2_imshow(img_out)
```

**Rotation, Shifting, and many more...**

```python
center = (img.shape[1]//2, img.shape[0]//2)
angle = 30
scale = 1

rotation_mat = cv.getRotationMatrix2D(center, angle, scale)
rotated_img = cv.warpAffine(img, rotation_mat, (img.shape[1], img.shape[0]))

cv2_imshow(rotated_img)
```

```python
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
kernel = np.ones((3, 3), np.uint8)

dilated = cv.dilate(img_gray, kernel, iterations=2)
eroded = cv.erode(dilated, kernel, iterations=2)
opening = cv.morphologyEx(eroded, cv.MORPH_OPEN, kernel)
closing = cv.morphologyEx(opening, cv.MORPH_CLOSE, kernel)

cv2_imshow(dilated)
cv2_imshow(eroded)
cv2_imshow(opening)
cv2_imshow(closing)
```

```python
for gamma in [0.1, 0.5, 1.2, 2.2]:
  gamma_corrected = np.array(255 * (img / 255) ** gamma, dtype='uint8')
  cv.imwrite('gamma_transformed' + str(gamma) + '.jpg', gamma_corrected)
```

```python
down = cv.pyrDown(img)
up = cv.pyrUp(down)

cv2_imshow(down)
cv2_imshow(up)
```
