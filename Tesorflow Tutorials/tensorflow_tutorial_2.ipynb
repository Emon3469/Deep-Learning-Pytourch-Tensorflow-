```python
import tensorflow as tf
from tensorflow import keras
import kagglehub
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('/content/sonar_dataset.csv', header=None)
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)
```

# **Dropout Regularization - To Tackle Overfitting**

```python
model = keras.Sequential([
    keras.layers.Dense(60, input_dim=60, activation='relu'),
    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

y_train_encoded = y_train.map({'R': 0, 'M': 1})
y_test_encoded = y_test.map({'R': 0, 'M': 1})

model.fit(X_train, y_train_encoded, epochs=100, batch_size=8)
```

```python
model.evaluate(X_test, y_test_encoded)
```

```python
y_pred = model.predict(X_test)
print(y_pred[:10])

y_pred = np.round(y_pred)
print(y_pred[:10])
```

```python
from sklearn.metrics import confusion_matrix , classification_report

print(classification_report(y_test_encoded, y_pred))
```

```python
modeld = keras.Sequential([
    keras.layers.Dense(60, input_dim=60, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1, activation='sigmoid')
])

modeld.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

modeld.fit(X_train, y_train_encoded, epochs=100, batch_size=8)
```

```python
modeld.evaluate(X_test, y_test_encoded)
```

```python
y_pred = modeld.predict(X_test).reshape(-1)
y_pred = np.round(y_pred)
print(y_pred[:10])
```

```python
print(classification_report(y_test_encoded, y_pred))
```

```python
y_train.value_counts(), y.value_counts()
```

```python
def ANN(X_train, y_train, X_test, y_test, loss, weights):
  model = keras.Sequential([
    keras.layers.Dense(27, input_dim=27, activation='relu'),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
  ])
  model.compile(loss=loss, optimizer='adam', metrics=['accuracy'])

  if weights == -1:
    model.fit(X_train, y_train, epochs=100)
  else:
    model.fit(X_train, y_train, epochs=100, class_weight = weights)
  print(model.evaluate(X_test, y_test))

  y_preds = model.predict(X_test)
  y_preds = np.round(y_preds)

  print("Classification Report: \n", classification_report(y_test, y_preds))

  return y_preds
```

```python
y_preds = ANN(X_train, y_train, X_test, y_test, 'binary_crossentropy', -1)
```

# **Ensemble with undersampling**

```python
X = df2.drop('Churn', axis='columns')
y = df2['Churn']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=15, stratify=y)
```

```python
df3 = X_train.copy()
df3['Churn'] = y_train
df3.head()
```

```python
df3_class0 = df3[df3.Churn==0]
df3_class1 = df3[df3.Churn==1]

def get_train_batch(df_majority, df_minority, start, end):
    df_train = pd.concat([df_majority[start:end], df_minority], axis=0)

    X_train = df_train.drop('Churn', axis='columns')
    y_train = df_train.Churn
    return X_train, y_train

X_train, y_train = get_train_batch(df3_class0, df3_class1, 0, 1495)
y_pred2 = ANN(X_train, y_train, X_test, y_test, 'binary_crossentropy', -1)
```

```python
X_train, y_train = get_train_batch(df3_class0, df3_class1, 1495, 2990)
y_pred1 = ANN(X_train, y_train, X_test, y_test, 'binary_crossentropy', -1)
```

```python
X_train, y_train = get_train_batch(df3_class0, df3_class1, 2990, 4130)
y_pred3 = ANN(X_train, y_train, X_test, y_test, 'binary_crossentropy', -1)
```

```python
y_pred_final = y_pred1.copy()
for i in range(len(y_pred1)):
  n_ones = y_pred1[i] * y_pred2[i] * y_pred3[i]
  if n_ones > 1:
    y_pred_final[i] = 1
  else:
    y_pred_final[i] = 0

c1_rep = classification_report(y_test, y_pred_final)
c1_rep
```
